```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def meme_gen(image, text_up=str, text_down=str):
    img = cv2.imread(image)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    font = cv2.FONT_HERSHEY_COMPLEX
    text_size = cv2.getTextSize(text_up, font, 1, 5)[0]
    textx = (img.shape[1]-text_size[0]) // 2
    texty = (img.shape[0]+text_size[1]) // 2
    img1 = cv2.putText(img, text_up, (textx+50, texty-160), font, 0.85, (255, 255, 255), 2)
    meme = cv2.putText(img1, text_down, (textx+80, texty+150), font, 1, (255, 255, 255), 2)
    meme = cv2.cvtColor(meme, cv2.COLOR_RGB2BGR)
    cv2.imwrite('meme.jpg', meme)
    plt.figure(figsize=(10, 20))
    plt.imshow(meme)

meme_gen('//Users/szokirov/Downloads/Futurama-Fry.jpeg', 'Not sure if this is working correctly', 'Or I am just making it up')
```
